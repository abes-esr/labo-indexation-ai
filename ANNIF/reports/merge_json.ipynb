```python
import glob
import json
import pandas as pd

file_names = glob.glob('./*-fr.json')

json_list = {}

for curr_f_name in file_names:
    name = str(curr_f_name.split("/")[-1])
    with open(curr_f_name) as curr_f_obj:
        json_list[name] = json.load(curr_f_obj)

with open('merged_annif_results_concepts_chaines.json', 'w') as out_file:
    json.dump(json_list, out_file, indent=4)
```

```python
# Convert to df
annif_eval = pd.read_json(
    'merged_annif_results_concepts_chaines.json',
    orient="index")
```
